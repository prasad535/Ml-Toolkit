# Why visualization

- Let assume we have 1000K records, its difficult identify the insights to make the decisions.
- Visualization helps to make good decision by using data records.

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.random.rand(50)
y = np.random.rand(50)
```

```python
x
```

```python
y
```

```python
plt.scatter(x, y)
```

```python
plt.scatter(x,y)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('this is X vs Y')
```

```python
plt.scatter(x, y, c='b')
```

```python
plt.scatter(x, y, c='#F315DF')
```

```python
plt.scatter(x, y, c='#F315DF', alpha=0.3)
```

```python
plt.scatter(x, y, c='#F315DF', alpha=0.5)
plt.grid()
```

```python
plt.figure(figsize=(3,3))
plt.scatter(x, y, c='#F315DF', alpha=0.5, marker='x')
plt.grid()
```

```python
x = np.linspace(1, 10, 100)
y = np.sin(x)
```

```python
x
```

```python
y
```

```python
plt.plot(x, y, c='r')
```

```python
plt.scatter(x, y)
```

```python
plt.plot(x, y, '--r')
```

```python
plt.plot(x, y, 'r+')
```

```python
plt.figure(figsize=(6,6))
plt.plot(x, y, 'bo')
plt.title('Sing graph')
```

```python
x = ['a', 'b', 'c', 'd', 'e']
y = np.random.rand(5)
```

```python
plt.bar(x, y)
plt.grid()
```

```python
plt.bar(x, y, color = "#E112CE")
plt.grid()
```

```python
plt.barh(x, y, color = "#12E13B")
plt.grid()
```

```python
x = [3,4,5,6,7,8,9]
y = [4,5,6,7,8,9,10]

plt.figure(figsize=(6,2))
plt.plot(x,y)
```

```python
x = np.random.rand(50)
y = np.random.rand(50)
z = np.random.rand(50)

fig = plt.figure()
ax = fig.add_subplot(projection = '3d')
ax.scatter(x,y,z)
plt.show()
```
